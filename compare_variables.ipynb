```python
from utils import adstock, s_curve, create_transformed_tables, plot_stacked_area_with_actual, plot_transformed_vs_raw, plot_data_set
import pandas as pd

df_values = pd.read_csv('Data/df_values.csv', index_col=0, parse_dates=True)
df_values2 = pd.read_csv('Data/df_values2.csv', index_col=0, parse_dates=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(df_values.index, df_values['media1'], label='media1 (1)', color='red')
plt.plot(df_values2.index, df_values2['media1'], label='media1 (2)', color='blue')

plt.xlabel('Date')
plt.ylabel('media1 value')
plt.title('media1 Comparison: df_data vs df_values')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(df_values.index, df_values['actual'], label='actual(1)', color='red')
plt.plot(df_values2.index, df_values2['actual'], label='actual (2)', color='blue')

plt.xlabel('Date')
plt.title('Sales Comparison')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate media1_pct for both dataframes
df_values['media1_pct'] = df_values['media1'] / df_values['actual']
df_values2['media1_pct'] = df_values2['media1'] / df_values2['actual']

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df_values.index, df_values['media1_pct'], label='media1_pct (1)', color='red')
plt.plot(df_values2.index, df_values2['media1_pct'], label='media1_pct (2)', color='blue')
plt.xlabel('Date')
plt.ylabel('media1 / actual')
plt.title('media1_pct Comparison')
plt.legend()
plt.tight_layout()
plt.show()
```
